```python
# 1. Download and read the data file
```

```python
import pandas as pd
```

```python
coupon = pd.read_excel("http://byungwan.com/class/Coupon_Assignment.xlsx")
```

```python
coupon.head()
```

```python
# 2(a). A bar chart and a pie chart of CClass
```

```python
from matplotlib import pyplot as plt
```

```python
coupon["CClass"].value_counts()
```

```python
cclass_freq = coupon["CClass"].value_counts().values
# make an array -> contains 2 values
```

```python
cclass_freq
```

```python
cclass_index = coupon["CClass"].value_counts().index
```

```python
plt.figure(figsize=(3,2)) # 3 is width, 2 is height
plt.bar(cclass_index, cclass_freq)
plt.show()
```

```python
# Bar Chart
plt.figure(figsize=(3,2))
plt.bar(cclass_index, cclass_freq)
plt.xlabel("cclass", fontsize=12)
plt.ylabel("Count (persons)", fontsize=12)
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(4,4))
plt.pie(cclass_freq, labels=["Category 2", "Category 1", "Category 3", "Category 0"], autopct="%.2f%%",explode=(0, 0, 0.5, 0))
plt.show()
```

```python
# 2.(b) A bar chart and a histogram of Sales
```

```python
plt.figure(figsize=(2,3))
plt.boxplot(coupon["Sales"], tick_labels=["Sales"], notch=True)
plt.show()
```

```python
plt.figure(figsize = (3,3))
plt.hist(coupon["Sales"], bins=20)
plt.show()
```

```python
# 2.(c) A scatter chart of Sales and Discount
```

```python
plt.figure(figsize=(4,2))
plt.scatter(x=coupon["Sales"], y = coupon["Discount"], s = coupon["Visits"]*10, alpha=0.5)
plt.ylabel("Discount", fontsize=12)
plt.xlabel("Sales", fontsize=12)
plt.margins(0.1, 0.3)
plt.grid(color="lightgray")
plt.show()
```

```python
# 2.(d) A heatmap of correlation matrix of all variables
```

```python
import seaborn as sns
```

```python
corrs = coupon.corr()
corrs
```

```python
plt.figure(figsize = (7,7))
sns.heatmap(corrs, annot=True)
plt.show()
```
