### Visualization of Categorical Variable 
- Bar chart
- Pie chart

## 1. Bar chart

```python
from matplotlib import pyplot as plt
```

```python
coupon["Gender"].value_counts()
```

```python
gender_freq = coupon["Gender"].value_counts().values
# make an array -> contains 2 values
```

```python
gender_index = coupon["Gender"].value_counts().index
```

```python
gender_freq
```

```python
gender_index
```

```python
plt.figure(figsize=(3,2)) # 3 is width, 2 is height
plt.bar(gender_index, gender_freq)
plt.show()
```

```python
# Bar Chart
plt.figure(figsize=(3,2))
plt.bar(gender_index, gender_freq, color="pink")
plt.xticks([0,1], ["Male", "Female"]) # only 0 and 1, got rid of 0.5
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Count (persons)", fontsize=12)
plt.show()
```

## 2. Pie chart

```python
plt.figure(figsize=(2,2))
plt.pie(gender_freq, labels=["Male", "Female"]) # 소수점 아래 몇자리까지 보여줄 것인지 작성
plt.show()
```

```python
plt.figure(figsize=(2,2))
plt.pie(gender_freq, labels=["Male", "Female"], autopct="%.2f%%", colors=["lightblue", "coral"]) # 소수점 아래 몇자리까지 보여줄 것인지 작성
plt.show()
```

```python
plt.figure(figsize=(2,2))
plt.pie(gender_freq, labels=["Male", "Female"], autopct="%.2f%%", colors=["lightblue", "coral"],explode=(0.1, 0))
plt.show()
```

```python
plt.figure(figsize=(2,2))
plt.pie(gender_freq, labels=["Male", "Female"], autopct="%.2f%%", colors=["lightblue", "coral"],explode=(0.5, 0))
plt.show()
```

### Visualization of Numerical Variables

### 3. Box chart

```python
coupon.head()
```

```python
plt.figure(figsize=(2,3))
plt.boxplot(coupon["Sales"], tick_labels=["Sales"])
plt.show()
```

### 4. Histogram

```python
plt.figure(figsize = (2,2))
plt.hist(coupon["Sales"])
plt.show()
```

```python
plt.figure(figsize = (2,2))
plt.hist(coupon["Sales"], bins=50)
plt.show()
```

```python
plt.figure(figsize = (3,2))
plt.hist(coupon["Sales"], bins="auto", facecolor="lightblue")
plt.xlabel("Sales", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

### Visualization of Numerical Variables
- Scatter chart -> show how they might be related
- visualize multiple variables in one figure

```python
plt.figure(figsize=(4,2))
plt.scatter(x=coupon["Sales"], y = coupon["Gender"])
plt.show()
```

```python
plt.figure(figsize=(4,2))
plt.scatter(x=coupon["Sales"], y = coupon["Gender"])
plt.yticks([0,1], ["Male","Female"])
plt.ylabel("Gender", fontsize=12)
plt.xlabel("Sales", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(4,2))
plt.scatter(x=coupon["Sales"], y = coupon["Gender"], s = coupon["Visits"]*20, alpha=0.5) # alpha is adding transparency
plt.yticks([0,1], ["Male","Female"])
plt.ylabel("Gender", fontsize=12)
plt.xlabel("Sales", fontsize=12)
plt.margins(0.1, 0.3)
plt.grid(color="lightgray")
plt.show()
```

```python
# we can see if there is a tendency or not -> there aren't any trend
```

```python
## in this scatter plot -> 3 variables
# Sales, Gender, Visits
# We can add the color of the scatter plot too
```

```python
import numpy as np
```

```python
colormap = np.array(["lightblue","coral"])
```

```python
plt.figure(figsize=(4,2))
plt.scatter(x=coupon["Sales"], 
            y = coupon["Gender"], 
            s = coupon["Visits"]*20, 
            alpha=0.5, 
            c = colormap[coupon["Discount"]]) # discount has 0, 1 as value
plt.yticks([0,1], ["Male","Female"])
plt.ylabel("Gender", fontsize=12)
plt.xlabel("Sales", fontsize=12)
plt.margins(0.1, 0.3)
plt.grid(color="lightgray")
plt.show()

# s means size, c means color

# 0 -> did not receive discount, 1 -> has received discount
```
