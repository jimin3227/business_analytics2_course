### 1. Simple linear Regression

```python
house2.head()
```

```python
house2.columns.get_loc("Price")
# get the column number
```

```python
house2.columns.get_loc("SqFt")
```

```python
y = house2.iloc[:, 1]
# price is what we use as y variable
# : -> means all rows, column 1
```

```python
x = house2.iloc[:, 2]
```

```python
# linear regression can be found at statsmodels.api
import statsmodels.api as sm
```

```python
reg = sm.OLS(y, x).fit()
```

```python
print(reg.summary())
```

```python
# Regression analysis finds the equation of the best-fitting line that minimizes the sum of the squares of all the observed errors
# OLS : Ordinary Least Squares
```

```python
# coef for SqFt is 65.2384 : slope of the linear regression
# beta minimize the sum of square of errors --> OLS
```
