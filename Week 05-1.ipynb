## Missing Values

```python
loan.shape
```

```python
loan.isnull().values.any()
```

```python
loan.isnull().sum()
```

## 1. Mean substitution

```python
loan["LoanAmount"].describe()
```

```python
### 614개 중에 592개밖에 없고 그들의 평균 146
```

```python
LA_Mean = loan["LoanAmount"].mean()
LA_Mean
```

```python
loan.fillna({"LoanAmount": LA_Mean}, inplace=True)
```

```python
loan.isnull().sum()
```

```python
loan["LoanAmount"].describe()
```

```python
# 카운트 늘어있어야 정상
```

## 2. Mode substitution numeric 이 아니라 categorical 이라면...

```python
loan["Self_Employed"].value_counts()
```

```python
loan["Self_Employed"].mode()
# mode는 최빈값
```

```python
SF_Mode = loan["Self_Employed"].mode()[0]
SF_Mode
```

```python
# 이러면 no가 제거됨
```

```python
loan.fillna({"Self_Employed": SF_Mode}, inplace=True)
```

```python
loan.isnull().sum()
```

```python
loan["Self_Employed"].value_counts()
```

```python
# 여전히 mode 는 no 임.
```
