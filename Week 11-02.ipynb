### Neural Network

### 0. Data

```python
import pandas as pd
```

```python
coupon = pd.read_excel("http://byungwan.com/class/Coupon_Sample.xlsx")
coupon.head()
```

```python
coupon["Discount"].value_counts()
```

```python
from sklearn.utils import resample
```

```python
coupon0 = coupon[coupon["Discount"] == 0] 
```

```python
coupon1 = coupon[coupon["Discount"] == 1]
```

```python
coupon0_downsample = resample(coupon0, replace=False, n_samples=122, random_state=0)
```

```python
coupon_downsample = pd.concat([coupon0_downsample, coupon1], axis=0)
```

```python
coupon_downsample.head()
```

```python
coupon_downsample["Discount"].value_counts()
```

```python
y = coupon_downsample.iloc[:, 4]
x = coupon_downsample.iloc[:, [1,2,3,5,6,7,8,9,10]]
```

```python
x.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler_model = scaler.fit(x)
```

```python
scaled_x = scaler_model.transform(x)
```

```python
pd.DataFrame(scaled_x, columns=x.columns)
```

```python
pd.DataFrame(scaled_x, columns=x.columns).describe()
```

## 1. ANN

```python
from sklearn.neural_network import MLPClassifier
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
nums = list(range(1, 11))
nums
```

```python
for n in nums:
    mlp_n = MLPClassifier(activation='relu', hidden_layer_sizes=(n), max_iter=3000, random_state=0)
    scores = cross_val_score(mlp_n, scaled_x, y, cv=5)
    mean = np.mean(scores)
    print(str(n)+ ": "+ str(mean))
```

```python
# alternative is to chose the highest accuracy score, but the first highest score that is found will be selected
# first time that we can get the highest is 5 nodes 
```

```python
mlp = MLPClassifier(activation='relu', hidden_layer_sizes=(5), max_iter=3000, random_state=0)
```

```python
mlp_model = mlp.fit(scaled_x, y)
```

#### 1.1. Evaluation

```python
coupon_test = pd.read_excel("http://byungwan.com/class/Coupon_Test.xlsx")
coupon_test.head()
```

```python
y_test = coupon_test.iloc[:, 4]
x_test = coupon_test.iloc[:, [1,2,3,5,6,7,8,9,10]]
```

```python
scaled_x_test = scaler_model.transform(x_test)
```

```python
y_pred = mlp_model.predict(scaled_x_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test, y_pred)
```

```python
# why is the accuracy lower on the test set?
'''
posssible reasons: 1) overfitting

2) test data set Discount value counts may be very unbalanced compared to training set

There might be bias from the training data


'''
```

```python
coupon_test["Discount"].value_counts()
```
