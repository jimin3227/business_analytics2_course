#### Support Vector Machine

#### 0.Data

```python
import pandas as pd
```

```python
house = pd.read_excel("http://byungwan.com/class/House_Prices2.xlsx")
house.head()
```

```python
y = house.iloc[:, 6]
x = house.iloc[:, [2,3,4,5,7,8,9]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

### 1. Linear SVM

```python
from sklearn.svm import SVC
```

```python
lsvm = SVC(kernel='linear')
```

```python
lsvm_model = lsvm.fit(x_train, y_train)
```

```python
y_pred = lsvm_model.predict(x_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test, y_pred)
```

### 2. Non-linear SVM

```python
nlsvm = SVC(kernel='rbf')
```

```python
nlsvm_model = nlsvm.fit(x_train, y_train)
```

```python
y_pred = nlsvm_model.predict(x_test)
```

```python
metrics.accuracy_score(y_test, y_pred)
```

### 2.1. Parameter tuning

```python
nlsvm2 = SVC(kernel='rbf', C=100 , gamma= 0.01)
# C value is whether we are allowing some misclassification or not
## smaller C value means we are allowing some observation within the margin
# C = 100 is sort of hard margin / high C = hard margin
# gamma is how much curvature we are allowing
## larger gamma means more curvature
```

```python
nlsvm_model2 = nlsvm2.fit(x_train, y_train)
```

```python
y_pred = nlsvm_model2.predict(x_test)
```

```python
metrics.accuracy_score(y_test, y_pred)
```

### 2.2. Grid Search

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
params = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
          'gamma': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
```

```python
grid = GridSearchCV(SVC(kernel='rbf'), params, cv=5)
```

```python
grid_model = grid.fit(x, y)
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```
