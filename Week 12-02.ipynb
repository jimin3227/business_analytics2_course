## Support Vector Machine

### 0.Data

```python
import pandas as pd
```

```python
coupon = pd.read_excel("http://byungwan.com/class/Coupon_Sample.xlsx")
coupon.head()
```

```python
coupon["Discount"].value_counts()
```

```python
y = coupon.iloc[:, 4]
# columns except 4th column
x = coupon.iloc[:, [1,2,3,5,6,7,8,9,10]]
```

### 1. Non-linear SVM

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
params = {"C": [0.001, 0.01, 0.1, 1, 10, 100, 1000],
          "gamma": [0.001, 0.01, 0.1, 1, 10, 100, 1000],
          "kernel": ['rbf','sigmoid']}
```

```python
grid = GridSearchCV(SVC(), params, cv=5)
```

```python
grid_model = grid.fit(x,y)
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```

### 1.1. Best model

```python
svc = SVC(kernel="rbf", C=100, gamma=0.01)
```

```python
svc_model = svc.fit(x,y)
```

#### 1.2. Evaluation

```python
coupon_test = pd.read_excel("http://byungwan.com/class/Coupon_Test.xlsx")
coupon_test.head()
```

```python
coupon.head()
```

```python
y_test = coupon_test.iloc[:, 4]
x_test = coupon_test.iloc[:, [1,2,3,5,6,7,8,9,10]]
```

```python
y_pred = svc_model.predict(x_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test, y_pred)
```

### 2. Linear SVM

```python
params = {"C": [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
```

```python
grid = GridSearchCV(SVC(kernel='linear'), params, cv=5)
```

```python
grid_model = grid.fit(x,y)
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```
